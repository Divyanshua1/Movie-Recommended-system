```python
movies = movies.merge(credits,on='title')
```

```python
movies.dropna(inplace = True)
```

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(genre):
  L =[]
  for i in ast.literal_eval(genre):
    L.append(i['name'])
    return L
```

```python
movies['genres'] =movies['genres'].apply(convert)
```

```python
movies.head(5)
```

```python
movies['keywords'] =movies['keywords'].apply(convert)
```

```python
def fetch_director(obj1):
  L =[]
  for i in ast.literal_eval(obj1):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(5)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split()) # to convert overview into list
```

```python
movies.head(5)
```

```python
def convert1(obj2):
  L=[]
  for i in ast.literal_eval(obj2):
    L.append(i['name'])
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
movies.head(5)
```

```python
movies['tag'] = movies['overview'] +movies['cast'] + movies['crew'] +movies['genres']
```

```python
new_df =movies[['movie_id','title','tag']]
movies.head(5)
```

```python
import pandas as pd

# Assuming 'new_df' is your DataFrame with a 'tag' column containing lists and possibly some non-list values

# Define a function to convert lists to strings
def convert_to_string(x):
    if isinstance(x, list):
        return ' '.join(map(str, x))
    else:
        return str(x)

# Apply the function to the 'tag' column
new_df['tag'] = new_df['tag'].apply(convert_to_string)
```

```python
new_df.head(5)
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 ,stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
            y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loving')
```

```python
ps.stem('dance')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity  = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True ,key = lambda x:x[1])[1:6]
```

```python
def recommended(movie):
        movies_index = new_df[new_df['title'] == movie].index[0]
        distances = similarity[movies_index]
        movies_list = sorted(list(enumerate(distances)),reverse = True ,key = lambda x:x[1])[1:6]
        for i in movies_list:
            print(new_df.iloc[i[0]].title)
```

```python
recommended('Avatar')
```
